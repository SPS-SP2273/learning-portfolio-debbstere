## 1 A simple plot

```python
from matplotlib import pyplot as plt
```

### 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.plot(x, y_2, color='red', linestyle='dashed', linewidth=2)
plt.show()
plt.plot(x, y_2, c='red', ls='dashed', lw=2)
plt.show()

# ^ Produce the same graph
# you can use c --> color, ls --> linestyle, le --> linewidth

# Don't forget to type plt.show() to produce the graph
```

```python
# the default settings for plotting graphs

plt.plot(x, y_1)
plt.show()
```

### 1.2 Adding another plot

```python
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')     # Second plot
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()

# You can plot 2 sets of data on the same graph by adding another plt.plot() line below the previous before the plt.show()
# command.
```

### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, color='green', label='Y3 with errors')    # Third plot w/ error bars
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='best')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='right')

plt.show()
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.array(np.linspace(0, 50, 1000))
y = np.e**(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x', fontsize=15, color='darkblue')
plt.ylabel('y', rotation =0, fontsize=15, color='darkblue')
plt.grid(alpha=0.25)
plt.ylim(-1, 1)
plt.title(r'$y = e^{-x/10}sin(x)$', fontsize=20)
plt.show()
```
