## 1 Comparing the two ‘dialects’

```python
from matplotlib import pyplot as plt
```

### 1.1 Some nomenclature

A figure is the full area we use for plotting our diagrams and such. An axis is an individual mathematical axis used for plotting

Commands such as `plt.twinx()` allows us to plot a graph with multiple axes.

### 1.2 We need some data.

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### 1.3 Here comes the comparison

```python
# The usual method of generating a plot

plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
plt.show()
```

```python
# Using the OO interface for plotting

fig, ax = plt.subplots(nrows=1, ncols=1)     # Need to first create subplots which will give us a figure and an axis
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
plt.show()
```

## 2 Using the OO Interface

### 2.1 Getting ax

```python
fig, ax = plt.subplots(ncols=1, nrows=2)

# Gives us 2 axes to carry out the plotting on
```

### 2.2 What is ax

```python
print(type(ax))       # The axis created is a numpy array
print(ax.shape)       # The axis created contains 2 'items'
print(type(ax[0]))    # The axis is a matplotlib axis
print(type(ax[1]))    # Is the same, a matplotlib axis because they are the same
```

### 2.3 Plots in a column

```python
fig, ax = plt.subplots(ncols=1, nrows=2)
ax[0].plot(x, cos_x, label='cos x')   # Should assign cos(x) to the first axis located at the top
ax[1].plot(x, sin_x, label='sin x')   # Should assign sin(x) to the second axis located at the bottom
plt.show()
```

### 2.4 Legends

```python
ax[0].legend()
ax[1].legend()

# To create legends for each plot, we need to call each axis. To do this more efficiently for any number of axes, we can
# use a for loop:

for a in ax:
    a.legend()

# Adding a grid to each axis
    
for a in ax:
    a.legend()
    a.grid(alpha=.25)
```

### 2.5 Tweaks

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 10),   # 10 x 5 inches!
                       sharex=True)

# Specifying the figure size and have the plots share the same x-axis

ax[1].set_xlabel('$x$')

# Setting labels for the x-axis
# Note that we have to use set_xlabel() rather than xlabel() for the OO interface

ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].fill_between(x, 0, sin_x, alpha=.25)

# Fills the areas between the curves with a colour of 25% opacity

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

# Add a supertitle at the top of all the graphs 
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(10, 5),   # 10 x 5 inches!
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

# Place all the axis specifications together each after each axis

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

plt.show()

# Without tight layout
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(10, 5),   # 10 x 5 inches!
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

# Place all the axis specifications together each after each axis

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()

plt.show()

# With tight layout
```

## 3 More rows and columns

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10, 4),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

# x data for plotting
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(10, 20, 100)

top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
plt.show(ax.flatten)
```

### 3.1 Using ax

```python
ax[0, 0].plot(x1, np.ones_like(x1))
ax[0, 1].plot(x2, 2*np.ones_like(x2))
ax[1, 0].plot(x1, 3*np.ones_like(x1))
ax[1, 1].plot(x2, 4**np.ones_like(x2))

# The way to access a 2 by 2 figure

top_left=ax[0, 0]
top_right=ax[0, 1]
bottom_left=ax[1, 0]
bottom_right=ax[1, 1]

# The first index refers to the plots numbered from top to bottom (column number). The second index refers to the plots
# numbered from left to right (row number)

top_left, top_right, bottom_left, bottom_right = ax.flatten()

# You can also use unpacking with ax.flatten() which takes a 2D array and flattens it into a 1D array
```

### 3.2 Sharing axes

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10, 4))

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

plt.show()

# Unshared axes will produce plots with separate axes for all plots since matplotlib will auto-scale all the plots
```

Methods to specify how to share the axes:

| Option | Result |
|:------:|:------:|
|`True`|Makes **all** the axes use the same range|
|`col`|Uses the same range for all the columns|
|`row`|Uses the same range for all the rows|

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10, 4),
                       sharex=True, sharey='row')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

plt.show()

# All plots share the same range for the x-axis and plots in the same row share the same axis for the y-axis
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(10, 4),
                       sharex='col', sharey='col')

top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.plot(x1, np.ones_like(x1))
top_right.plot(x2, 2*np.ones_like(x2))
bottom_left.plot(x1, 3*np.ones_like(x1))
bottom_right.plot(x2, 4**np.ones_like(x2))

plt.show()

# All plots in the same column will share the same x and y axis
```

### 3.3 Accessing all axes

```python
top_left.grid(alpha=.25)
top_right.grid(alpha=.25)
bottom_left.grid(alpha=.25)
bottom_right.grid(alpha=.25)

# Applying changes to all the axes

for a in ax.flatten():
    a.grid(alpha=.25)
    
# More efficient way to apply changes to all the axes
```

## 4 Other useful plots

### 4.1 Histograms

```python
data = {}

filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
data['All'] = np.loadtxt(filename, skiprows=1)

filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
data['NUS'] = np.loadtxt(filename, skiprows=1)

# Loading the data in the .csv files into the dictionary data

plt.style.use('bmh')

plt.hist([data['All'], data['NUS']],
         bins=50,                        # How many bins to split the data
         label=['All', 'NUS']
         )
plt.xlabel('Mean of Basic Montly Eraning (S$)')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

### 4.2 Scatter plots

```python
data = {}
for label in ['All', 'NUS']:
    filename = f'sg-gov-graduate-employment-survey_basic_monthly_mean_{label.lower()}.csv'
    data[label] = np.loadtxt(filename, skiprows=1)

# Using a for loop to load the data

plt.style.use('seaborn-darkgrid')

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)
    
plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
plt.show()
```

### 4.3 Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('ggplot')
plt.bar(majors, numbers)      # plt.bar() for vertical bars and plt.barh() for horizontal bars
plt.xlabel('Majors')
plt.ylabel('Number of Students')

plt.show()
```

### 4.4 Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=45        # Changes the angle in which the pie chart is created in        
        )
plt.title('Percentage of each major')
plt.show()

# For formatting percentages: autopct= , input has to be a str. The first % indicates that the value should be in
# percentages. You can choose to add or not add the first 1 but .1 or .2 means round to 1 decimal place or 2 decimal
# places respectively. .0 means round to the nearest whole number. Without the .<num>, just a number will display the
# percentages as a float as how it is calculated. im not sure what the letter does but you can use other letters like
# e to display the percentage in scientific notation. If you want to display the % symbol, you must add %% behind the
# letter.
```

## Exercise 1 :  Plotting data from files

```python
plt.style.use('default')

#--------- Generate cosine and sine values --------#
x = np.linspace(-np.pi, np.pi, num=100, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=2, ncols=2,
                         figsize=(12, 8),  sharey='row', sharex='col')

#------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label='$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r',
                linestyle=':', label='$\cos^2 x$')
axes[0, 0].set_title('$\cos x$ & $\cos^2x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].fill_between(x, cos_x, -1, color='r', alpha=.125)
axes[0, 0].legend(loc='upper right')
axes[0, 0].set_xlim(-np.pi, np.pi)

#------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label='$\sin x$')
axes[0, 1].set_title('$\sin x$ & $\sin^2x$')
axes[0, 1].fill_between(x, sin_x, -2, color='g', alpha=.125)
axes[0, 1].plot(x, sin_x**2, label='$\sin^2 x$')
axes[0, 1].set_ylim(-1.25, 1.25)
axes[0, 1].legend(loc='lower right', frameon=False)
axes[0, 1].legend(loc='upper right')
axes[0, 1].set_xlim(-np.pi, np.pi)

#------- Subplot 3 -------#
axes[1, 0].plot(x, fun1_x, color='b', label='$e^{-x}\cos 5x$')
axes[1, 0].fill_between(x, fun1_x, -25, color='b', alpha=.125)
axes[1, 0].set_title('$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
axes[1, 0].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 0].set_xticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])
axes[1, 0].legend(loc='upper right')

#------- Subplot 4 -------#
axes[1, 1].plot(x, fun2_x, color='y', label='$e^{-x}\sin 2x$')
axes[1, 1].set_title('$e^{-x}\sin 2x$')
axes[1, 1].fill_between(x, fun2_x, -25, color='y', alpha=.125)
axes[1, 1].set_xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
axes[1, 1].set_xticklabels(['$-\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$'])
axes[1, 1].set_xlabel('Angle (radians)')
axes[1, 1].legend(loc='upper right')

# 'flatten', 'opens' the 2D array into a simple 1D array
for a in axes.flatten():
    a.grid(alpha=.5)

plt.tight_layout()    # Makes all the labels and plots more compact and nicely spaced out
plt.show()
```
