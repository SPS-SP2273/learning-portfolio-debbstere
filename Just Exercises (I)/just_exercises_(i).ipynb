## Exercise 4 :  A function for Fibonacci

```python
def fibonacci(n):
    fib = []
    a, b = 0, 1
    for i in range(n):
        fib.append(a)
        a, b = b, (a+b)
    return fib

fibonacci(15)
```

## Exercise 5 :  max_info() without NumPy

```python
def max_info(numbers):
    max_info = numbers[0]
    for i, num in enumerate(numbers):
        if num > max_info:
            max_info = numbers[i]
    return max_info, numbers.index(max_info)

numbers = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]

max_info(numbers)
```
